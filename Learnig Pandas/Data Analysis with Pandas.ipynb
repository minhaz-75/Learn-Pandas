Linear Interpolation:

Definition: Linear interpolation is the simplest form of interpolation. It approximates values between two known data points using a straight line.
Difference: Linear interpolation assumes a linear relationship between data points, meaning the estimated values lie on a straight line connecting the neighboring data points.
When to use: Linear interpolation is suitable when the data points show a relatively consistent trend, and the underlying relationship between data points is approximately linear.

y_interpolated = y1 + (x_interpolation - x1) * (y2 - y1) / (x2 - x1)


Barycentric Interpolation:

Definition: Barycentric interpolation is an alternative to polynomial interpolation that uses weighted averages of data points to approximate the values in between.
Difference: Barycentric interpolation can provide a more stable and accurate interpolation for equidistant data points compared to standard polynomial interpolation.
When to use: Barycentric interpolation is useful when dealing with equidistant data points and when stability is crucial in the interpolation process.


PCHIP (Piecewise Cubic Hermite Interpolating Polynomial):

Definition: PCHIP is a type of cubic spline interpolation that uses piecewise cubic polynomials to approximate the data.
Difference: PCHIP guarantees monotonicity, meaning it preserves the order of the data points, and produces a smoother curve compared to some other interpolation methods.
When to use: PCHIP is particularly useful when the data shows monotonic behavior, and preserving monotonicity is essential for the application.

Akima Interpolation:

Definition: Akima interpolation is a type of spline interpolation that uses a weighted average of slopes to estimate intermediate values.
Difference: Akima interpolation is designed to handle noisy data and avoids oscillations or wild fluctuations in the interpolated curve.
When to use: Akima interpolation is suitable for dealing with noisy data or when avoiding large oscillations in the interpolated curve is essential.

y_interpolated = y_i + B(x - x_i)


Spline Interpolation:

Definition: Spline interpolation involves fitting a piecewise polynomial function to the data points, creating a smooth curve through the data.
Difference: Spline interpolation is flexible and can be implemented with various polynomial degrees, such as linear, quadratic, or cubic splines, depending on the level of smoothness desired.
When to use: Spline interpolation is commonly used when the data shows significant variation and a smooth curve is required to capture the overall trend.

S(x) = Si(x), where x_i <= x < x_(i+1)


Polynomial Interpolation:

Definition: Polynomial interpolation approximates the data using a single polynomial equation, passing through all the given data points.
Difference: Polynomial interpolation can result in a high-degree polynomial, leading to oscillations (Runge's phenomenon) and poor accuracy if the degree is too high or the data points are unevenly spaced.
When to use: Polynomial interpolation can be useful for low-degree polynomials or when the data points are well-behaved and evenly spaced. However, caution is needed when using high-degree polynomials, especially with sparse or noisy data.

P(x) = a0 + a1*x + a2*x^2 + ... + an*x^n

# Pivot table

We generally work on pivot table provided in Excel, which is very handy to
get a detailed summary of the data and analyse the data. Pandas provide the

```python
weather_pivot = {
    'date':['01/03/20','01/03/20','01/03/20','01/03/20','01/03/20','01/03/20','12/01/19','12/01/19','12/01/19','12/01/19','12/01/19','12/01/19'],
    'city':['mumbai','mumbai','delhi','delhi','kolkata','kolkata','mumbai','mumbai','delhi','delhi','kolkata','kolkata'],
    'temperature':[43,42,40,42,38,37,22,20,18,19,21,23],
    'windspeed':[9,11,8,8,6,8,12,10,9,7,7,10],
    'time':['morning','evening','morning','evening','morning','evening','morning','evening','morning','evening','morning','evening',]
}

df = pd.DataFrame(weather_pivot)
df.to_csv('weather_pivotTable.csv',index=False)
df = pd.read_csv('weather_pivotTable.csv')
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df.pivot_table(index='city', columns='date')
```

Custom functions to individual column
